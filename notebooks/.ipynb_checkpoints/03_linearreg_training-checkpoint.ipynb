## Linear Regression Models


```python
import pandas as pd
import numpy as np

X_train= pd.read_csv(r"C:\Users\LENOVO\Desktop\project_ml_retail\data\train_test\others\X_train_scaled.csv")
X_test= pd.read_csv(r"C:\Users\LENOVO\Desktop\project_ml_retail\data\train_test\others\X_test_scaled.csv")
y_train= pd.read_csv(r"C:\Users\LENOVO\Desktop\project_ml_retail\data\train_test\others\y_train.csv")
y_test= pd.read_csv(r"C:\Users\LENOVO\Desktop\project_ml_retail\data\train_test\others\y_test.csv")


print("shape:",X_train.shape)
X_train.head()
```

```python
y_train = y_train.squeeze()  # (3497,1) → (3497,)
y_test  = y_test.squeeze()   # (874,1)  → (874,)
```

```python
print(f"y_train shape : {y_train.shape}")
print(f"y_train type  : {type(y_train)}")
print(y_train.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

# Train
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict
y_pred_train = lr.predict(X_train)
y_pred_test  = lr.predict(X_test)

# Evaluate
def evaluate_model(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    print(f"\n=== {name} ===")
    print(f"R²   : {r2:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"MAE  : {mae:.4f}")

evaluate_model("Train", y_train, y_pred_train)
evaluate_model("Test",  y_test,  y_pred_test)
```


Train R² = 0.3952 → model explains 39.5% of spending variance
Test  R² = 0.4157 → model explains 41.5% of spending variance

Meaning:
Out of all the reasons why customers spend
different amounts, your features (habits +
demographics) explain about 40% of it

The other 60% is explained by factors
you don't have:
→ Product category bought
→ Promotions/discounts applied
→ Random personal decisions
→ Economic conditions

LinearRegression → baseline, no constraints
Ridge            → reduces overfitting, keeps all features
Lasso            → reduces overfitting, can eliminate features

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score

models = {
    'LinearRegression': LinearRegression(),
    'Ridge (L2)'      : Ridge(alpha=1.0),
    'Lasso (L1)'      : Lasso(alpha=0.1),
}

results = []

for name, model in models.items():
    # Cross validation on train
    cv_scores = cross_val_score(
        model, X_train, y_train,
        cv=5, scoring='r2'
    )

    # Fit and evaluate on test
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2   = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae  = mean_absolute_error(y_test, y_pred)

    results.append({
        'Model'     : name,
        'CV R² Mean': cv_scores.mean().round(4),
        'CV R² Std' : cv_scores.std().round(4),
        'Test R²'   : round(r2, 4),
        'Test RMSE' : round(rmse, 4),
        'Test MAE'  : round(mae, 4)
    })
    print(f"{name}: CV R²={cv_scores.mean():.4f} "
          f"(±{cv_scores.std():.4f}) | Test R²={r2:.4f}")

results_df = pd.DataFrame(results)
print("\n=== MODEL COMPARISON ===")
print(results_df.to_string(index=False))
```

LinearRegression ≈ Ridge ≈ Lasso

This tells you:
✅ No overfitting (Ridge didn't improve)
✅ Features are not highly correlated
   (Ridge didn't help)
✅ All features carry some signal
   (Lasso didn't zero out many)
✅ Linear relationships exist in data

```python
from sklearn.linear_model import RidgeCV, LassoCV

# RidgeCV — automatically finds best alpha
ridge_cv = RidgeCV(
    alphas=[0.01, 0.1, 1, 10, 100, 1000],
    cv=5
)
ridge_cv.fit(X_train, y_train)
print(f"Best Ridge alpha: {ridge_cv.alpha_}")

# LassoCV — automatically finds best alpha
lasso_cv = LassoCV(
    alphas=[0.001, 0.01, 0.1, 1, 10],
    cv=5, max_iter=5000
)
lasso_cv.fit(X_train, y_train)
print(f"Best Lasso alpha: {lasso_cv.alpha_}")

# Evaluate both
for name, model in [('RidgeCV', ridge_cv), 
                     ('LassoCV', lasso_cv)]:
    y_pred = model.predict(X_test)
    evaluate_model(name, y_test, y_pred)
```

```python
import matplotlib.pyplot as plt

# Use best model — likely Ridge
best_model = ridge_cv

coef_df = pd.DataFrame({
    'Feature'    : X_train.columns,
    'Coefficient': best_model.coef_
})

coef_df['Abs_Coef'] = coef_df['Coefficient'].abs()
coef_df = coef_df.sort_values('Abs_Coef', ascending=False)

print("=== TOP 15 MOST IMPORTANT FEATURES ===")
print(coef_df.head(15).to_string(index=False))

# Plot
plt.figure(figsize=(12, 8))
top15 = coef_df.head(15)
colors = ['green' if x > 0 else 'red'
          for x in top15['Coefficient']]

plt.barh(top15['Feature'], top15['Coefficient'],
         color=colors)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.title('Top 15 Feature Coefficients — Ridge Regression')
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
```
